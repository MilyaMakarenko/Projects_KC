````python
taxi_peru = pd.read_csv('3_taxi_peru.csv', sep=';', parse_dates=['start_at', 'end_at', 'arrived_at'])
````

````python
taxi_filter = taxi_peru.query('(start_type == "asap") | (start_type == "reserved")')
````

````python
taxi_filter['wait_time'] = (taxi_filter.arrived_at - taxi_filter.start_at).astype('timedelta64[m]')
````

````python
taxi_filter.query('(start_type == "reserved") & wait_time > 0.0') \
    .groupby('driver_id') \
    .agg({'journey_id': 'count', 'driver_score': 'mean', 'rider_score': 'mean'}) \
    .sort_values('journey_id', ascending=False)
````

````python
taxi_peru['month'] = taxi_peru.start_at.dt.month
````

````python
taxi_peru['weekday'] = taxi_peru.start_at.dt.strftime('%A')
````

````python
sns.countplot(taxi_peru.month)
````

````python
sns.countplot(taxi_peru.weekday, order=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
````

````python
taxi_month = taxi_peru.groupby('month', as_index=False) \
    .agg({'user_id': 'nunique'})
````
